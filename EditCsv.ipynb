```python
import pandas as pd
import csv

file_path = r'C:\\Users\\doubl\\Desktop\\김경민\\2023\\Multicampus\\파이널\\test\\column.csv'
new_file_path = r'C:\\Users\\doubl\\Desktop\\김경민\\2023\\Multicampus\\파이널\\test\\column_fix.csv'

data = []
with open(file_path, 'r', newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        try:
            data.append(row)
        except csv.Error:
            continue

df = pd.DataFrame(data)
for idx, row in df.iterrows():
    for col in range(4, 9):  # 4~8열 확인
        if not pd.isnull(row[col]) and 'https' not in row[col]:  # 만약 값이 있고 'https'를 포함하지 않는 경우
            df.iloc[idx, col:] = df.iloc[idx, col:].shift(1)  # 해당 열부터 모든 열을 오른쪽으로 한 칸 이동

df.to_csv(new_file_path, index=False)
```

```python
import pandas as pd
import csv

file_path = r'C:\\Users\\doubl\\Desktop\\김경민\\2023\\Multicampus\\파이널\\test\\new_test.csv'
new_file_path = r'C:\\Users\\doubl\\Desktop\\김경민\\2023\\Multicampus\\파이널\\test\\column_fix_2.csv'

data = []
with open(file_path, 'r', newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        try:
            data.append(row)
        except csv.Error:
            continue

df = pd.DataFrame(data)

# 각 행의 본문과 이미지를 분리
for idx, row in df.iterrows():
    # 이미지 링크가 시작되는 열 찾기
    img_start_col = next((i for i, x in enumerate(row[3:], start=3) if pd.notnull(x) and 'https' in x), None)
    if img_start_col is not None:
        # 본문 열 이동
        df.iloc[idx, 3:img_start_col] = df.iloc[idx, 3:img_start_col].shift(3-img_start_col)
        # 이미지 열 이동
        df.iloc[idx, img_start_col:] = df.iloc[idx, img_start_col:].shift(8-img_start_col)

df.to_csv(new_file_path, index=False)
file_path = r'C:\\Users\\doubl\\Desktop\\김경민\\2023\\Multicampus\\파이널\\test\\new_test.csv'
new_file_path = r'C:\\Users\\doubl\\Desktop\\김경민\\2023\\Multicampus\\파이널\\test\\column_fix_3.csv'
```

```python
import pandas as pd
import numpy as np

# csv 파일 읽기
df = pd.read_csv(r'C:\\Users\\doubl\\Desktop\\csv\new_test_2.csv', header=None)

# 키워드 설정
keywords = ["정보", "특징", "맛과 어울리는 음식", "양조장 정보"]

# 새로운 DataFrame 생성
new_df = pd.DataFrame(columns=range(len(df.columns)))

for idx in range(1, len(df)):
    # 술 이름 가져오기
    name = df.iloc[idx, 0]
    # 새로운 행 생성
    new_row = [''] * len(df.columns)
    # 첫 번째 칼럼에 술 이름 채우기
    new_row[0] = name
    # 각 키워드에 대해 처리
    for i in range(len(keywords)):
        # 키워드와 술 이름 조합
        keyword = f'{name} {keywords[i]}'
        # 키워드가 포함된 열의 인덱스 찾기
        keyword_cols = [j for j, x in enumerate(df.iloc[idx]) if pd.notnull(x) and keyword in x]
        if keyword_cols:
            start_col = keyword_cols[0]
            # 다음 키워드가 포함된 열의 인덱스 찾기
            if i+1 < len(keywords):
                next_keyword = f'{name} {keywords[i+1]}'
                next_keyword_cols = [j for j, x in enumerate(df.iloc[idx]) if pd.notnull(x) and next_keyword in x]
                if next_keyword_cols:
                    end_col = next_keyword_cols[0]
                else:
                    end_col = None
            else:
                end_col = None
            # 데이터 채우기
            if end_col is not None:
                new_row[start_col:end_col] = df.iloc[idx, start_col:end_col]
            else:
                new_row[start_col:] = df.iloc[idx, start_col:]
    # 새로운 행 추가
    new_df.loc[len(new_df)] = new_row

# csv 파일로 저장
new_df.to_csv(r'C:\\Users\\doubl\\Desktop\\csv\\column_fix_3.csv', index=False)
```

```python
import pandas as pd
import numpy as np

# CSV 파일 읽기
df = pd.read_csv(r'C:\\Users\\doubl\\Desktop\\csv\\new_test.csv')

# 각 카테고리별 최대 갯수를 저장할 빈 딕셔너리 생성
category_max = {'정보': 0, '특징': 0, '음식': 0, '양조장': 0}

# 각 술마다 정보, 특징, 음식, 양조장 데이터의 갯수를 확인하여 최대값 업데이트
for i in range(len(df)):
    row = df.iloc[i, :].tolist()
    name, *contents = row
    contents = [content.strip() for content in contents if not pd.isna(content)]
    
    info_start = next((i for i, content in enumerate(contents) if f'노란수트다빈치 정보' in content), None)
    feature_start = next((i for i, content in enumerate(contents) if f'노란수트다빈치 특징' in content), None)
    food_start = next((i for i, content in enumerate(contents) if f'노란수트다빈치 맛과 어울리는 음식' in content), None)
    brewery_start = next((i for i, content in enumerate(contents) if f'노란수트다빈치 양조장 정보' in content), None)
    
    info_count = feature_start - info_start if feature_start is not None and info_start is not None else (len(contents) - info_start if info_start is not None else 0)
    feature_count = food_start - feature_start if food_start is not None and feature_start is not None else (len(contents) - feature_start if feature_start is not None else 0)
    food_count = brewery_start - food_start if brewery_start is not None and food_start is not None else (len(contents) - food_start if food_start is not None else 0)
    brewery_count = len(contents) - brewery_start if brewery_start is not None else 0
    
    category_max['정보'] = max(category_max['정보'], info_count)
    category_max['특징'] = max(category_max['특징'], feature_count)
    category_max['음식'] = max(category_max['음식'], food_count)
    category_max['양조장'] = max(category_max['양조장'], brewery_count)

# 각 카테고리별 최대 갯수를 기준으로 새로운 데이터프레임 생성
new_df = pd.DataFrame(columns=['이름'] + ['정보']*category_max['정보'] + ['특징']*category_max['특징'] + 
                              ['음식']*category_max['음식'] + ['양조장']*category_max['양조장'])

# 기존 데이터프레임의 각 행을 새로운 데이터프레임에 맞게 재배열
for i in range(len(df)):
    row = df.iloc[i, :].tolist()
    name, *contents = row
    contents = [content.strip() for content in contents if not pd.isna(content)]
    
    info_start = next((i for i, content in enumerate(contents) if f'노란수트다빈치 정보' in content), None)
    feature_start = next((i for i, content in enumerate(contents) if f'노란수트다빈치 특징' in content), None)
    food_start = next((i for i, content in enumerate(contents) if f'노란수트다빈치 맛과 어울리는 음식' in content), None)
    brewery_start = next((i for i, content in enumerate(contents) if f'노란수트다빈치 양조장 정보' in content), None)
    
    info_contents = contents[info_start:feature_start] if info_start is not None and feature_start is not None else (contents[info_start:] if info_start is not None else [])
    feature_contents = contents[feature_start:food_start] if feature_start is not None and food_start is not None else (contents[feature_start:] if feature_start is not None else [])
    food_contents = contents[food_start:brewery_start] if food_start is not None and brewery_start is not None else (contents[food_start:] if food_start is not None else [])
    brewery_contents = contents[brewery_start:] if brewery_start is not None else []
    
    info_contents.extend([np.nan]*(category_max['정보'] - len(info_contents)))
    feature_contents.extend([np.nan]*(category_max['특징'] - len(feature_contents)))
    food_contents.extend([np.nan]*(category_max['음식'] - len(food_contents)))
    brewery_contents.extend([np.nan]*(category_max['양조장'] - len(brewery_contents)))
    
    new_row = [name] + info_contents + feature_contents + food_contents + brewery_contents
    new_df.loc[i] = new_row

# 새로운 데이터프레임을 CSV 파일로 저장
new_df.to_csv(r'C:\\Users\\doubl\\Desktop\\csv\\column_fix_3.csv', index=False)
```
