```python
# 서울_매실원주_15도_크롤링

from bs4 import BeautifulSoup
import requests
import csv
import os

# 웹페이지 요청
url = "https://terms.naver.com/entry.naver?docId=3595250&cid=58637&categoryId=58644"
res = requests.get(url)
res.raise_for_status()

# 웹페이지 파싱
soup = BeautifulSoup(res.text, 'lxml')

# 술 제목 추출
name = soup.find('h2', class_='headword').get_text(strip=True)  

# 술 소제목 추출
stress_1 = soup.find('h3', class_='stress').get_text(strip=True)  # "매실원주 정보" 추출
stress_2 = soup.find(id='TABLE_OF_CONTENT2').get_text(strip=True)  # "매실원주 특징" 추출
stress_3 = soup.find(id='TABLE_OF_CONTENT7').get_text(strip=True)   # 매실원주 15도 맛과 어울리는 음식
stress_4 = soup.find(id='TABLE_OF_CONTENT8').get_text(strip=True)   # 매실원주 양조장 정보

# 술 소제목 번호 추출
num_1 = soup.find(id='TABLE_OF_CONTENT3').get_text(strip=True)  # "1. 좋은 재료, 좋은 사람, 좋은 제품" 추출
num_2 = soup.find(id='TABLE_OF_CONTENT4').get_text(strip=True)  # "2. 변함 없는 재료에 대한 철학" 추출
num_3 = soup.find(id='TABLE_OF_CONTENT5').get_text(strip=True)  # "3. 높은 주세를 감수하고 지켜는 품질" 추출
num_4 = soup.find(id='TABLE_OF_CONTENT6').get_text(strip=True)   # "4. 국내외 주류품평회 수상" 추출
num_5 = soup.find(id='TABLE_OF_CONTENT9').get_text(strip=True)  # "1. 매실원주 생산자" 추출
num_6 = soup.find(id='TABLE_OF_CONTENT10').get_text(strip=True)  # "2. 매실원주 생산지" 추출

# 술 설명 추출
txt_1 = soup.find('p', class_='txt').get_text(strip=True)  # 설명 추출
txt_2 = soup.find_all('p', class_='txt')[1].get_text(strip=True)  # 두 번째 'p' 태그의 본문 추출, 더한주류는~부터 시작
txt_3 = soup.find_all('p', class_='txt')[2].get_text(strip=True)  # 세 번째 'p' 태그의 본문 추출, 매실원주는~부터 시작
txt_4 = soup.find_all('p', class_='txt')[3].get_text(strip=True)  # 네 번째 'p' 태그의 본문 추출, 우리나라~부터 시작
txt_5 = soup.find_all('p', class_='txt')[4].get_text(strip=True)  # 대한민국~부터 시작
txt_6 = soup.find_all('p', class_='txt')[5].get_text(strip=True)  # 매실원주, 300ml~부터 시작
txt_7 = soup.find_all('p', class_='txt')[6].get_text(strip=True)  # 맑은 담황색의~부터 시작
txt_8 = soup.find_all('p', class_='txt')[7].get_text(strip=True)  # 전통주는 고리타분~부터 시작
txt_9 = soup.find_all('p', class_='txt')[8].get_text(strip=True)  # 서울 은평구~부터 시작

# 이미지 추출
image_divs = soup.find_all('div', class_='thmb c')  # 이미지가 있는 모든 div 태그를 찾음
image_urls = [div.find('img')['src'] for div in image_divs]  # 각 div 태그에서 이미지 URL 추출

# csv파일로 저장
with open(os.path.join('C:\\Users\\doubl\\Desktop\\test', 'data_detail.csv'), 'w', newline='', encoding='utf-8-sig') as f:
    writer = csv.writer(f)
    writer.writerow([name, stress_1, txt_1, image_urls[0], image_urls[1], stress_2, num_1, image_urls[2], txt_2, num_2, image_urls[3], txt_3, num_3, image_urls[4], txt_4, num_4, image_urls[5], txt_5, stress_3, txt_6, image_urls[6], image_urls[7], txt_7, stress_4, num_5, image_urls[8], txt_8, num_6, image_urls[9], txt_9])
```
